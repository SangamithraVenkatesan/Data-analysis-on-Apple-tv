```python


# Load the dataset
df = pd.read_csv('weather_BE.csv')

# Encode the target variable
df['Rain'] = df['Rain'].map({'no rain': 0, 'rain': 1})

# Prepare features and target
X = df.drop('Rain', axis=1)
y = df['Rain']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize Logistic Regression model
model = LogisticRegression()

# Perform 10-fold cross-validation
cv_scores = cross_val_score(model, X_scaled, y, cv=10, scoring='accuracy')

# Print accuracy scores in percentage
print("Logistic Regression Accuracy Scores (10-fold):")
for i, score in enumerate(cv_scores, 1):
    print(f"Fold {i}: {score * 100:.2f}%")

# Print average accuracy
print(f"\nAverage Accuracy: {cv_scores.mean() * 100:.2f}%")
```
